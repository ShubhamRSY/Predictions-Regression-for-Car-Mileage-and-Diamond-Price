### Importing Required Libraries


```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## Load the data in a csv file into a dataframe


```python
# the data set is available at the url below.
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/mpg.csv"

# using the read_csv function in the pandas library, we load the data into a dataframe.

df = pd.read_csv(URL)
```


Let's examine a few example entries from the dataset we have imported:


```python
# show 8 random rows from the dataset
df.sample(8)
```

Let's find out the number of rows and columns in the dataset:


```python

# Counting the rows and columns in the dataset
rows = df.shape[0]
columns = df.shape[1]

print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")
```

To visualize the relationship between Horsepower and Mileage (MPG), we will create a scatter plot. This will allow us to observe how these two variables interact with each other.

You are encouraged to create more plots to visualize relationships amongst other columns


```python
import matplotlib.pyplot as plt

plt.scatter(df['Horsepower'], df['MPG'])
plt.xlabel('Horsepower')
plt.ylabel('Miles Per Gallon (MPG)')
plt.title('Scatter Plot of Horsepower vs. MPG')
plt.show()
```

##  Determine the Dependent Variable and the Independent Variables



First, we determine the prediction goal. This is the value that our machine learning model is designed to forecast.


```python
target = df["MPG"]
```

Next, we'll focus on identifying the features. Features are the data attributes that our machine learning model uses for learning and making predictions.

```python
features = df[["Horsepower","Weight"]]
```

 ### Build and Train a Linear Regression Model


Develop a linear regression model.


```python
lr = LinearRegression()
```

Train or fit the model.

```python
lr.fit(features,target)
```

## Assess the model and forecast outcomes.


It's time to assess the performance of our trained model.


```python
#Higher the score, better the model.
lr.score(features,target)
```

Let's forecast the fuel efficiency (mileage) for a car that has a horsepower of 100 and a weight of 2000 pounds.


```python
lr.predict([[100,2000]])
```

```python
## The fuel efficiency of a car with 100 horsepower and a weight of 2000 pounds is 29.3216098 miles per gallon.
```

# Regression for Diamond price


```python
URL2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/diamonds.csv"
```

###  Loading the dataset (diamond dataset)


```python
df2 = pd.read_csv(URL2)
```

```python
df2.sample(8)
```

```python
# Counting the rows and columns in the dataset
rows = df2.shape[0]
columns = df2.shape[1]

print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")
```

###  Determine the column to be predicted and the columns containing the data.


- Utilizing the 'price' column as the target variable.
- Employing 'carat' and 'depth' columns as the feature variables.

```python
target = df2["price"]
features = df2[["carat","depth"]]
```

### Construct and train a new linear regression model.

Constructing a new model for linear regression.


```python
lr2 = LinearRegression()
lr2.fit(features,target)
```

###  Evaluating the model


Obtain the model's rating.


```python
lr2.score(features,target)
```

###  Predicting the price of a diamond


Predict the price of a diamond with carat = 0.3 and depth = 60


```python
lr2.predict([[0.3, 60]])
```
